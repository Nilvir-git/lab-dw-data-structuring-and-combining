```python
pivot_df = df.pivot_table(index='sales_channel', columns='month', values=['monthly_premium_auto'], aggfunc='sum').round(2)

pivot_df.loc['Total'] = pivot_df.sum()

pivot_df['Total'] = pivot_df.sum(axis = 1)

pivot_df

#if we consider 'sales' to be 'monthly_premium_auto', the following table would show us the information that we need, top month sales is '1', and the best sales_channel is 'agent'.
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_df_clv = df.pivot_table(index='gender', columns='education', values=['customer_lifetime_value'], aggfunc='sum').round(2)

pivot_df_clv.loc['Total'] = pivot_df_clv.sum()

pivot_df_clv['Total'] = pivot_df_clv.sum(axis = 1)

pivot_df_clv

# in this case the top gender in customer_lifetime_value would be 'F', and based on education 'Bachelor' is slightly higher.
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
df.columns
```

```python
#policy_type
#month
#number_of_open_complaints
```

```python
complaints_summary = df.groupby(['policy_type', 'month'])['number_of_open_complaints'].sum().reset_index()

complaints_summary.columns = ['policy_type', 'month', 'num_complaints']

complaints_summary = complaints_summary.sort_values(by='num_complaints', ascending=False)

display(complaints_summary)
```
